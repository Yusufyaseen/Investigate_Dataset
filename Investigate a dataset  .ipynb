### Data Cleaning 

#### Drop id ,imdb_id,homepage ,tagline and overview columns

```python
tmdb.drop(columns=['id', 'imdb_id','homepage','tagline','overview'],inplace=True)
tmdb.info()
```

#### split  genres ,cast ,keywords and production_companies values  for visualization .

##### split genres values

```python
# load only non values 
genres_df = tmdb.loc[tmdb['genres'].notnull(),:]
# split each genre and assign it to data frame 
genres_df = genres_df.assign(genres=genres_df['genres'].str.split('|')).explode('genres')
#view data after editing 
genres_df.head()
```

##### split cast values

```python
# load only non values 
cast_df = tmdb.loc[tmdb['cast'].notnull(),:]
# split each cast member  and assign it to data frame
cast_df = cast_df.assign(cast=cast_df['cast'].str.split('|')).explode('cast')
# view data after editing 
cast_df.head()
```

##### split keywords values
